```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3, 5, 7, 6, 5, 67, 10, 15])
plt.plot(x, y, 'r')
plt.plot(x, y, 'g', lw=5) # line width
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3, 5, 7, 6, 5, 67, 10, 15])
plt.bar(x, y, 0.5, alpha = 1, color='b') #alpha is for png; para3: the width rate of bar
plt.show()
```
